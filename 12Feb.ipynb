```python
Q3.Which Python statements are used to catch and handle exceptions? Explain with an example
Ans:- we can use try and except statements to catch and handle exceptions. 

#Example

num1 = input("Enter a number: ")
num2 = input("Enter another number: ")

try:
    result = int(num1) / int(num2)
    print("The result is:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter a number.")
```

```python
Q4.
Ans:-

'''Try and else'''
the try statement is used to catch and handle exceptions that may occur in a block of code. 
The else statement is used to define a block of code that should be executed if no exception is raised in the try block. 

#example

try:                                      # code block that may raise an exception
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
except ValueError:
    print("Error: Invalid input. Please enter a number.")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
else:                                              # code block to execute if no exception is raised in the try block
    print("The result of division is:", result)
    
    
    
    
'''Finally'''                              # code block to execute irrespective of whether an exception is raised or not

try:
    file = open("example.txt", "r")
    file_contents = file.read()
    print(file_contents)
except FileNotFoundError:
    print("Error: File not found.")
finally:
    file.close()
    
here finally block is useful for tasks that need to be executed, even if an exception is raised, 
such as closing a file, releasing resources, or cleaning up temporary files.




'''raise'''
    the raise keyword is used to raise an exception explicitly. We can use the raise keyword to raise an exception 
    in response to a specific error condition that we detect in our code. 

     raise ExceptionType("Error message")
        
        #example
        
def divide(num1, num2):
    if num2 == 0:
        raise ZeroDivisionError("Cannot divide by zero.")
    else:
        return num1 / num2

try:
    result = divide(10, 0)
    print(result)
except ZeroDivisionError as error:
    print(error)
    
    i understand in this example - the divide function takes two arguments num1 and num2 and returns the result of dividing 
    num1 by num2. If num2 is equal to zero, we raise a ZeroDivisionError using the raise keyword.
```

```python
Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.
Ans:- 
Custom exceptions in Python are user-defined exceptions that can be raised in response to specific error conditions in our code.
We can create custom exceptions by defining a new class that inherits from the built-in Exception class or any other built-in exception class.

We need custom exceptions to provide more specific information about the errors that occur in our code. 
#Example-  

we are building a banking application, we may want to raise a InsufficientFundsError exception when a user tries to
withdraw more money than their account balance. This custom exception provides more specific information about the error condition,
making it easier to handle and debug errors in our code



class InsufficientFundsError(Exception):
    pass

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError("Insufficient funds.")
        else:
            self.balance -= amount
            return self.balance

try:
    account = BankAccount(1000)
    balance = account.withdraw(1500)
    print("New balance:", balance)
except InsufficientFundsError as error:
    print("Error:", error)
```
